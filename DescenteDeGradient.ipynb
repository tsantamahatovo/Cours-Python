```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv(r"C:\Users\tsant\Downloads\Data.csv",sep=";")
```

```python
#USA: extraction des valeurs dans un array numpy pour faciliter le from scratch
y=data['USA'].values
```

```python
#création des valeurs X et Y pour l'AR(1): prediction de y(t) en utilisant y(t-1)
Y=y[1:]#valleur de t=1 à n
X=y[:-1]#valeur de t=0 à n-1
```

```python
#Initialisation des paramètres
#On commence à 0
phi0=0.0
phi1=0.0

#Réglages des hyperparamètres
learning_rate=0.001 #taille du pas
n_iterations=1000 #combien de fois on répète l'ajustement
n=len(Y)
```

```python
# Initialisation
phi0 = 0.0
phi1 = 0.0
learning_rate = 0.001
epochs = 1000

for i in range(epochs):
    # 1. Prédiction (Y_chapeau = phi0 + phi1 * X)
    Y_pred = phi0 + phi1 * X

    # 2. Calcul de l'erreur (MSE)
    error = Y - Y_pred

    # 3. Calcul des Gradients
    # On calcule comment bouger phi0 et phi1 pour réduire l'erreur
    d_phi0 = (-2/n) * np.sum(error)
    d_phi1 = (-2/n) * np.sum(error * X)

    # 4. Maj des paramètres
    phi0 = phi0 - (learning_rate * d_phi0)
    phi1 = phi1 - (learning_rate * d_phi1)

    # Affichage du progrès
    if i % 100 == 0:
        mse = np.mean(error**2)
        print(f"Itération {i}: MSE = {mse:.5f}")

print(f"\nRésultats finaux : phi0 = {phi0:.4f}, phi1 = {phi1:.4f}")
```
